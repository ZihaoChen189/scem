```r
# dnorm(): 正态分布的密度函数, 某个具体数值的概率密度. 例如, dnorm(x, mean = 0, sd = 1) 会计算标准正态分布在 x 点的概率密度
# pnorm(): 正态分布的累积分布函数, 随机变量在某个值以下的概率. 例如, pnorm(q, mean = 0, sd = 1) 用于计算标准正态分布中随机变量小于或等于 q 的概率
# qnorm(): 正态分布的分位数函数, 使累积分布函数等于特定概率的点的值. 例如, qnorm(p, mean = 0, sd = 1) 会计算标准正态分布中 p 分位数的值
# rnorm(): 生成正态分布的随机数. 模拟符合特定均值和标准差的正态分布的数据. 例如, rnorm(n, mean = 0, sd = 1) 会生成 n 个服从标准正态分布的随机数
```

```r
student_t_confidence_interval<-function(sample,confidence_level){ 
    sample<-sample[!is.na(sample)] # remove any missing values 
    n<-length(sample) # compute sample size 
    mu_est<-mean(sample) # compute sample mean 
    sig_est<-sd(sample) # compute sample sd 
    alpha = 1-confidence_level # alpha from gamma 
    t<-qt(1-alpha/2,df=n-1) # get student t quantile 
    l=mu_est-(t/sqrt(n))*sig_est # lower 
    u=mu_est+(t/sqrt(n))*sig_est # upper 
    return(c(l,u)) 
} 
```

```r
num_trials <- 100000 
sample_size <- 30 
mu_0 <- 1 
sigma_0 <- 3 
alpha <- 0.05 
set.seed(0)

single_alpha_coverage_simulation_df <- 
    data.frame(trial=seq(num_trials)) %>% 
        mutate(sample=map(.x=trial,.f=~rnorm(n=sample_size,mean=mu_0,sd=sigma_0))) %>%  
        mutate(ci_interval=map(.x=sample, .f=~student_t_confidence_interval(.x,1-alpha)))%>% 
        mutate(cover=map_lgl(.x=ci_interval, .f=~((min(.x)<=mu_0)&(max(.x)>=mu_0))))%>% 
# compute interval length: 
        mutate(ci_length=map_dbl(.x=ci_interval, .f=~(max(.x)-min(.x))))

# estimate of coverage probability: 
        single_alpha_coverage_simulation_df %>% pull(cover) %>% mean() 

single_alpha_coverage_simulation_df
```

```r
set.seed(0)
num_trials <- 1000
sample_size <- 30 
min_alpha_size <- 0.01
max_alpha_size <- 0.9
alpha_size_inc <- 0.01
mu_0 <- 1 
sigma_0 <- 3 

simulation_df <- crossing(
    trial = seq(num_trials),
    alpha_size = seq(min_alpha_size, max_alpha_size, alpha_size_inc)
)
# simulation_df

simulation_df <- simulation_df %>%
    mutate(
        samples = pmap(list(trial, alpha_size), .f=~rnorm(sample_size, mean=mu_0, sd=sigma_0)),
        ci_interval=pmap(list(samples, alpha_size), .f=~student_t_confidence_interval(.x,1-.y)),
        cover=map_lgl(.x=ci_interval, .f=~((min(.x)<=mu_0)&(max(.x)>=mu_0))),
        ci_length=map_dbl(.x=ci_interval, .f=~(max(.x)-min(.x)))
    ) %>%
    group_by(alpha_size) %>%
    summarise(
        coverage_probability = mean(cover),
        avg_ci_length = mean(ci_length)   
    )

simulation_df
tail(simulation_df)
```
