```r
hal<-Hawks$Hallux
hal<-hal[!is.na(hal)]
hal
```

```r
outlier_val <- 100 
num_outliers <- 10 
corrupted_hal <- c(hal,rep(outlier_val,times = num_outliers))  # 在hal后面追加10个异常值, 数值大小是100
corrupted_hal
```

```r
mean(hal) 
mean(corrupted_hal) 
```

```r
num_outliers_vect <- seq(0, 1000)
means_vect <- c()
for(num_outliers in num_outliers_vect) {
    corrupted_hal <- c(hal, rep(outlier_val, times = num_outliers))
    means_vect <- c(means_vect, mean(corrupted_hal))
}
means_vect[1]
```

```r
num_outliers_vect <- seq(0, 1000)
medians_vect <- c()
for(num_outliers in num_outliers_vect) {
    corrupted_hal <- c(hal, rep(outlier_val, times = num_outliers))
    medians_vect <- c(medians_vect, median(corrupted_hal))
}
dim(medians_vect)
```

```r
num_outliers_vect <- seq(0, 1000)
t_means_vect <- c()
for(num_outliers in num_outliers_vect) {
    corrupted_hal <- c(hal, rep(outlier_val, times = num_outliers))
    t_means_vect <- c(t_means_vect, mean(corrupted_hal, trim = 0.1))
}
t_means_vect  # 正常不少
```

```r
df_means_medians <- data.frame(num_outliers = num_outliers_vect, mean = means_vect, t_mean = t_means_vect, median = medians_vect) 
```

```r
df_means_medians %>% 
    pivot_longer(!num_outliers, names_to = "Estimator", values_to = "Value") %>% 
    ggplot(aes(x=num_outliers,color=Estimator, linetype=Estimator,y=Value)) + geom_line()+xlab("Number of outliers") 
```

```r
num_outliers <- function(input_data) {
    q25 <- quantile(input_data, 0.25, na.rm = TRUE)
    q75 <- quantile(input_data, 0.75, na.rm = TRUE)
    IQR <- q75 - q25
    # amount <- sum((input_data > q75 + 1.5 * IQR) | (input_data < q25 - 1.5 * IQR), na.rm = TRUE)  # NA判断！！！
    # return (amount)  # 返回个数

    # 自己的方法 :)
    outlier <- input_data[(input_data > q75 + 1.5 * IQR) | (input_data < q25 - 1.5 * IQR)]
    outlier <- na.omit(outlier)
    return (length(outlier))  # 返回异常值的个数
}
num_outliers(c(0,40,60,185))
```

```r
Gen_Y_samples <- function(m, n) {
    Y_sample <- data.frame(index = seq(m)) %>%
    mutate(Y = map_dbl(index, ~ sum(Gen_X_numbers(n)) ))
    return (Y_sample)
}
Gen_Y_samples(5, 2) 
```

```r
samples_Y <- Gen_Y_samples(50000, 3)
ggplot(samples_Y, aes(Y)) + geom_bar() + theme_bw()
```

```r
set.seed(0) 
n <- 1000 
sample_X <- data.frame(U=runif(n)) %>% mutate(X=case_when( 
    (0<=U)&(U<0.25)~3, 
    (0.25<=U)&(U<0.5)~10, 
    (0.5<=U)&(U<=1)~0)) %>% 
pull(X)  # 使用 pull 函数从数据框中提取 X 列
mode(sample_X)
```

```r
sample_X_0310 <- function(alpha, beta, n){ 
    sample_X <- data.frame(U=runif(n)) %>% mutate(X=case_when( 
    (0<=U)&(U<alpha)~3, 
    (alpha<=U)&(U<alpha+beta)~10, 
    (alpha+beta<=U)&(U<=1)~0)) %>% 
    pull(X) 
    return (sample_X) 
} 
```

```r
n <- 10000 
alpha <- 1/2 
beta <- 1/10 
sample_X <- sample_X_0310(alpha,beta,n) 
mean(sample_X) 
```

```r
var(sample_X) 
```

```r
n = 100 
alpha = 1/10 
samples <- data.frame(beta = seq(0, 9/10, 0.01)) %>% 
    mutate( sample_X =  map(beta, ~sample_X_0310(alpha,.x,n) )) %>% 
    mutate( samplemean = map_dbl(sample_X, mean) ) %>% 
    mutate( Expectation = 3*alpha + 10*beta) 
```

```r
df <- samples %>% pivot_longer(cols=c(samplemean, Expectation),  
names_to = 'name', values_to = 'value') 
ggplot(df, aes(x=beta, y=value, color=name)) +  geom_point()  + theme_bw()  
```
