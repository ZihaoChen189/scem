```r
penguinsv2 %>%
  mutate(flipper_bill_ratio = flipper_length_mm/bill_length_mm)  # 创建一个新数据属性
```

```r
penguinsv2 %>% rename(f_l_m = flipper_length_mm)  # rename 重命名
```

```r
penguinsv2 %>% arrange(bill_length_mm)  # 升序排列
```

```r
penguinsv2 %>% arrange(desc(bill_length_mm))  # 降序排列
```

```r
penguinsv2 %>% 
  summarize(num_rows=n(),  # 统计数据集的行数 也就是有多少条记录
            avg_weight_kg=mean(body_mass_g/1000, na.rm=TRUE),  # na.rm = TRUE 忽略缺失值
            avg_flipper_bill_ratio=mean(flipper_length_mm/bill_length_mm, na.rm=TRUE))
```

```r
penguinsv2 %>% 
  group_by(species)  # 根据某些标准对数据框的行进行分组
```

```r
penguinsv2 %>%  # 巧妙的办法计算分组之后的统计数据
  group_by(species) %>%
  summarize(num_rows=n(), 
            avg_weight_kg=mean(body_mass_g/1000, na.rm=TRUE), 
            avg_flipper_bill_ratio=mean(flipper_length_mm/bill_length_mm, na.rm=TRUE))
```

```r
# 这段代码的作用是统计 penguinsv2 数据集中每列中缺失值(NA)的数量
Num_NAs <- penguinsv2 %>% summarize(species=sum(is.na(species)), 
                                    bill_length_mm=sum(is.na(bill_length_mm)), 
                                    body_mass_g=sum(is.na(body_mass_g)), 
                                    flipper_length_mm=sum(is.na(flipper_length_mm)))
print(Num_NAs) 
```

```r
# cross 允许对指定列集应用同样的操作, everything()对所有列进行操作, 
Num_NAs <- penguinsv2 %>% summarize(across(everything(), ~sum(is.na(.x))))  # 
print(Num_NAs)
# ~sum(is.na(.x)) 对选定的每一列进行操作
# ~ 表示匿名函数
# .x 代表当前列 sum(is.na(.x)) 的作用是统计每列中缺失值（NA）的数量
#  ~sum(is.na(.x)) is equivalent to function(x){(sum(is.na(x)))}
```

```r
# 对所有数据进行汇总
penguinsv2 %>% summarize(across(where(is.numeric), ~mean(.x, na.rm=TRUE)))
```

```r
penguinsv2 %>% 
  select(-bill_length_mm) %>%
  group_by(species) %>%
  summarize(across(where(is.numeric), ~mean(.x, na.rm=TRUE)))
```

```r
penguin_bill_lengths_df <- penguinsv2 %>%
  arrange(desc(bill_length_mm)) %>%
  select(species, bill_length_mm) 
penguin_bill_lengths_df
```
