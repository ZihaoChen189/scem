```r
library(tidyverse)
library(palmerpenguins)

peng_total <- penguins %>%  # prepare our data
    select(body_mass_g, flipper_length_mm, species) %>%
    filter(species != "Gentoo") %>%
    drop_na() %>%
    mutate(species = as.numeric(species == "Adelie"))
```

```r
# 分割训练集和测试集
num_total <- peng_total %>% nrow()  # penguin数据总数
num_train <- floor(num_total * 0.75)  # 训练样本数量
num_test <- num_total - num_train  # 测试样本数量

set.seed(1)  # 设置随机种子以保证可复现性
test_inds <- sample(seq(num_total), num_test)  # 随机抽取测试集索引
train_inds <- setdiff(seq(num_total), test_inds)  # 训练集索引

peng_train <- peng_total %>% filter(row_number() %in% train_inds)  # 训练数据
peng_test <- peng_total %>% filter(row_number() %in% test_inds)  # 测试数据

# 分离特征向量和标签
peng_train_x <- peng_train %>% select(-species)  # 训练集特征向量
peng_train_y <- peng_train %>% pull(species)  # 训练集标签

peng_test_x <- peng_test %>% select(-species)  # 测试集特征向量
peng_test_y <- peng_test %>% pull(species)  # 测试集标签
```
